```python
# Filter rows for matches labeled as 'Final'
final_matches = df[df['match_type'] == 'Final']

# Combine home and away teams to capture all participants
final_teams = pd.concat([
    final_matches[['home_team', 'season']].rename(columns={'home_team': 'team'}),
    final_matches[['away_team', 'season']].rename(columns={'away_team': 'team'})
])

# Count unique seasons for each team
team_season_counts = final_teams.groupby('team')['season'].nunique().sort_values(ascending=False)

# Display results
print("Number of IPL seasons in which each team participated in the finals:")
print(team_season_counts)

# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=team_season_counts.values, y=team_season_counts.index, palette='coolwarm')
plt.title('Number of IPL Finals Appearances by Team')
plt.xlabel('Number of Seasons')
plt.ylabel('Team')
plt.show()
```

```python
# Pivot the data for stacked bar plot
final_appearances = final_teams.groupby(['season', 'team']).size().unstack(fill_value=0)
```
